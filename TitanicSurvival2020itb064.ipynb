```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
ds=sns.load_dataset('titanic')
ds
```

```python
ds.drop(ds.iloc[:, 8:13], inplace=True, axis=1)
ds
```

```python
ds = pd.get_dummies(ds, columns = ['pclass', 'sex', 'embarked'])
print(ds)
```

```python
ds['age'] = ds['age'].replace(np.nan, ds['age'].mean())
print(ds)
```

```python
ds = ds[['age','sibsp','parch','fare','pclass_1','pclass_2','pclass_3', 'sex_female','sex_male','embarked_C','embarked_Q','embarked_S','survived']]
```

```python
print(ds)
```

```python
X=ds.iloc[:,0:12]
X
```

```python
Y=ds.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.125, random_state=42)
```

```python
X_test
```

```python
from keras import Model, Input
from keras.layers import Dense

inp = Input(shape=(12,))
l1 = Dense(8, activation='relu')(inp)
out = Dense(1, activation='sigmoid')(l1)
model = Model(inputs=inp, outputs=out)
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
model.fit(x=X_train, y=y_train, epochs=150, shuffle=True)
```

```python
from sklearn.metrics import accuracy_score

y_test=np.array(y_test)
X_test=np.array(X_test)

Y_pred1=model.predict(X_test).round()


score = accuracy_score(y_test, Y_pred1, normalize=True)
print("Accuracy of test data: ", score)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)

Y_pred2=model.predict(X_train).round()

score = accuracy_score(y_train, Y_pred2, normalize=True)
print("Accuracy of training data: ", score)
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def create_model():
    inp = Input(shape=(12,))
    l1 = Dense(8, activation='relu')(inp)
    out = Dense(1, activation='sigmoid')(l1)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32, verbose=0)
scores = cross_val_score(model, X, Y, cv=5)
print("Cross-validation scores:", scores)
print("Mean score:", scores.mean())
print("Standard deviation:", scores.std())
```

The standard deviation is not too large, which suggests that the model is not overfitting.
