```python
iris = iris.sample(frac=1)
iris
```

```python
iris1 = iris.iloc[:125,:]
iris2 = iris.iloc[125:,:]
print(iris1)
print(iris2)
```

```python
iris1 = iris1.reset_index()
iris2 = iris2.reset_index()
print(iris1)
print(iris2)
```
