```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs=np.linspace(0,19, 100)
ts = xs*1.1 + 1
m=0.1
c=2
lr=0.01
y=[]
for i in range(len(xs)):
    pred=m*xs[i]+c
    err=ts[i]-pred
    dc=err*lr
    dm=err*lr*xs[i]
    m+=dm
    c+=dc

y=m*xs+c
```

```python
plt.plot(xs,ts)
plt.plot(xs,y)
```

```python
xs=np.linspace(0,19, 100)
ts = xs*xs*1.1 + 2.2*xs + 10
m1=0.1
m2=1.3
c=2
lr=0.00001
y=[]
for i in range(len(xs)):
    pred=m1*xs[i]*xs[i]+c+m2*xs[i]
    err=ts[i]-pred
    dc=lr*err
    dm1=lr*err*xs[i]*xs[i]
    dm2=lr*err*xs[i]
    m1+=dm1
    c+=dc
    m2+=dm2

y=m1*xs*xs + m2*xs + c
```

```python
plt.plot(xs,ts)
plt.plot(xs,y)
```
