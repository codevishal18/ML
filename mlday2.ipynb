```python
x=[]
y=[]
for v in range (0,10):
  y.append (v*v*v)
  x.append (v)
plt.plot(x,y)
```

```python
x=[]
y=[]
for v in range (0,10):
  y.append (v*v*v)
  x.append (v)
plt.plot(x,y)
x1=[]
y1=[]
for v in range (0,10):
  y1.append (v*v)
  x1.append (v)
plt.plot(x1,y1)
```

```python
x=[]
y=[]
z=[]
for v in range (0,10):
  y.append (v*v*v)
  x.append (v)
  z.append (2.3*v+6)
plt.plot(x,y,label='first graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2 plots in one graph')
plt.plot(x,z,label='second graph')
plt.legend()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,1,5,2,6]
plt.plot(x,y,color='blue',linestyle='dotted',linewidth=4,marker='o',markerfacecolor='red',markersize=5)
plt.ylim(1,5)
plt.xlim(1,8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('customized graphs')
plt.show()
```
